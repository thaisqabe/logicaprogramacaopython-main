## média entre duas notas

```python
#entrada
p1 = float(input("Digite a primeira nota entre 0 e 10: "))
p2 = float(input("Digite a segunda nota entre 0 e 10: "))

#processamento
media = (p1+p2)/2

#saida
print("A média das notas é:", media)
```

## valor de peças vendidas e comissão

```python
#entrada
pu=float(input("Digite o valor de cada peça: "))
qv=float(input("Digite quantas peças foram compradas: "))

#processamento
vtp = pu*qv
comi = (vtp*5)/100

#saida
print("O valor total das vendas foi:", vtp)
print("O valor da comissão é:", comi)

```

## valor final do carro

```python
#entrada
vf=float(input("Digite o valor do carro: "))

#processamento
vfc=((vf*45)/100)+(vf)
vtotal=((vfc*28)/100)+(vfc)

#saida
print("O valor total do carro com o imposto e a porcentagem do distribuidor é:", vtotal)
```

## horas e minutos convertidos

```python
#entrada
h=(float(input("Digite a hora: ")))
m=(float(input("Digite os minutos: ")))

#processamento
mconvert=m/60
res=mconvert+h

#saida
print("As horas e minutos convertidos ficam:", res)
```

## valor com horas extras

```python
#entrada
ht=float(input("Digite a quantidade de horas trabalhadas: "))
mt=float(input("Digite a quantidade de minutos trabalhados: "))
qgt=float(input("Digite quanto você ganha por hora: "))

#processamento
htc=mt/60+ht
gtotal=qgt*htc
gthe=((gtotal*70)/100)+gtotal

#saida
print("O valor que você irá receber com as horas extras é:", gthe)
```

## dinheiro gasto com gasolina na viagem

```python
#entrada
kp=float(input("Digite quantos quilômetros você irá percorrer: "))
cg=float(input("Digite quantos quilômetros você percorre com 1 litro de gasolina: "))
pg=float(input("Digite quanto você paga em 1 litro de gasolina: "))

#processamento
lg=kp/cg
pfinal=lg*pg

#saida
print("O valor que você gastará com gasolina na ida da sua viagem será:", pfinal)
```

## velocidade média de um carro

```python
#entrada
ppkm=float(input("Digite o percurso percorrido em quilômetros: "))
hgp=float(input("Digite as horas gastas no percurso: "))
mgp=float(input("Digite os minutos gastos no percurso: "))

#processamento
tpg=mgp/60+hgp
vm=ppkm/tpg

#saida
print("A velocidade média em km/h é: ", vm)
```
